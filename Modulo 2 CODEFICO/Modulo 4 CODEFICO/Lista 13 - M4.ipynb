```python
Servidor.py

from socket import *

sk_server = socket(AF_INET, SOCK_DGRAM)
sk_server.bind(("64.225.8.163", 9676))

file = open("index.html", "r")

mensagem, endereco_cliente = sk_server.recvfrom(2048)

sk_server.sendto(file.read(2048).encode(), endereco_cliente)

```
