**Instru√ß√µes:**

**O servidor deve realizar as seguintes tarefas:**

*   Iniciar e aguardar conex√µes de clientes em um endere√ßo e porta espec√≠ficos.

*   Gerar dois n√∫meros inteiros aleat√≥rios no intervalo de 1 a 100, calcular a soma desses n√∫meros e enviar a pergunta ao cliente.

*   Receber a resposta do cliente.

*   Verificar se a resposta fornecida pelo cliente est√° correta. Se estiver correta, enviar a mensagem "Correto!" ao cliente. Caso contr√°rio, enviar a mensagem "Errado!"

**O cliente deve realizar as seguintes tarefas:**

*   Conectar-se ao servidor.

*   Receber a pergunta do servidor.

*   Solicitar ao usu√°rio a resposta da pergunta e envi√°-la de volta ao servidor.

*   Receber a resposta do servidor e exibi-la na tela.

*   Continuar a intera√ß√£o com o servidor at√© que o usu√°rio decida encerrar a conex√£o.

*   Certifique-se de que o cliente e o servidor estejam em execu√ß√£o em terminais separados.

```python
#SERVIDOR
from socket import *
#Importar a biblioteca

sk_server = socket(AF_INET, SOCK_DGRAM)
#Definir o tipo de endere√ßo e protocolo
sk_server.bind(("127.0.0.1", 1234))
#Iniciar o ip e porta do servidor

mensagem, endereco_cliente = sk_server.recvfrom(2048)
#Receber mensagem

sk_server.sendto(mensagem, endereco_cliente)
#Enviar mensagem
```

```python
#SERVIDOR
import random
from socket import *


sk_server = socket(AF_INET, SOCK_DGRAM)

sk_server.bind(("127.0.0.1", 1234))

while True:
    print("Esperando pergunta do cliente")
    mensagem, endereco_cliente = sk_server.recvfrom(2048)
    aleatorio1 = random.randint(1,100)
    aleatorio2 = random.randint(1,100)
    resposta = aleatorio1 + aleatorio2
    conta = str(aleatorio1) + " + " + str(aleatorio2)
    sk_server.sendto(conta.encode(), endereco_cliente)
    print("Esperando pergunta do cliente")
    mensagem, endereco_cliente = sk_server.recvfrom(2048)
    if(int(mensagem) == resposta):
        sk_server.sendto("Correto!!".encode(), endereco_cliente)
    else:
        sk_server.sendto("Incorreto!!".encode(), endereco_cliente)
```
