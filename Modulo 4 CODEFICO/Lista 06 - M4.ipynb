**Passo 3**: Crie um programa Python chamado `recalculate.py` para recalcular a pontua√ß√£o e salvar a tabela recalculada. Use um editor de texto para criar um arquivo chamado `recalculate.py` e adicione o seguinte c√≥digo:

```python
import csv

# Fun√ß√£o para recalcular a pontua√ß√£o total
def recalcular_pontuacao(row):
    vitorias = int(row['Vitorias'])
    empates = int(row['Empates'])
    derrotas = int(row['Derrotas'])
    return vitorias * 3 + empates - derrotas

# Leitura do arquivo CSV original
with open('tabela.csv', 'r') as arquivo:
    leitor = csv.DictReader(arquivo)
    cabecalho = leitor.fieldnames

    # Lista para armazenar as linhas recalculadas
    linhas_recalculadas = []

    for linha in leitor:
        # Recalcula a pontua√ß√£o e adiciona √† lista
        linha['Pontuacao'] = recalcular_pontuacao(linha)
        linhas_recalculadas.append(linha)

# Escrita do arquivo recalculado
with open('tabela_recalculada.txt', 'w') as arquivo_saida:
    # Escreve o cabe√ßalho
    arquivo_saida.write(','.join(cabecalho) + '\n')

    # Escreve as linhas recalculadas
    for linha in linhas_recalculadas:
        # Converter os valores inteiros em strings antes de escrever
        linha_str = [str(valor) for valor in linha.values()]
        arquivo_saida.write(','.join(linha_str) + '\n')

```

**Passo 4**: Crie um script Python chamado `check_existence.py` para verificar a exist√™ncia do arquivo `tabela-a_recalculada.txt`. Use um editor de texto para criar um arquivo chamado `check_existence.py` um exemplo do c√≥digo:

```python
import os
import subprocess

# Verifica se o arquivo tabela-a_recalculada.txt existe
if not os.path.exists('tabela_recalculada.txt'):
    # Se n√£o existe, execute o programa Python
    print("Rodando recalculate")
    subprocess.run(["python3", "recalculate.py"])

    # Ordene o resultado usando o comando sort com o Pipe e a op√ß√£o -n
    subprocess.run(["sort", "-t,", "-k5,5n", "tabela_recalculada.txt", "-o", "tabela_recalculada.txt"])
else:
    print("O arquivo tabela-a_recalculada.txt j√° existe.")


```
